# Goal : Save a dictionary to file .
## Deep Dive : Json.dump().JSON is the standard for data exchange.
## NOTE : JSON key must be strings;Python allows integers but JSON converts them.

The goal is to store a Python dictionary in a file so that it can be reused later.
To achieve this, we use **JSON (JavaScript Object Notation)** as the storage format.

---
### What is JSON?

JSON is a **standard data exchange format** used for:

* Saving structured data
* Transferring data between systems
* Storing configuration files
* Working with APIs and datasets

Python provides a built-in module called `json` to handle JSON operations.

---

## Important Note About JSON Keys

* In JSON, **all keys must be strings**
* Python dictionaries allow keys of any hashable type, including integers
* When converting Python dictionaries to JSON, integer keys are automatically converted to strings

### Example:

Python dictionary:

```python
{1: "one"}
```

JSON representation:

```json
{"1": "one"}
```

## Example 1: Dump (Dictionary → JSON File)

````python
import json
json.dump({'a':1} ,open('j.json','w'))
````

### Explanation:

* `{'a': 1}` is the Python dictionary
* `'j.json'` is the output file name
* `'w'` means write mode
* `json.dump()` writes the dictionary into a JSON file

### File content (`j.json`):

```json
{"a": 1}


## Example 2: Load (JSON File → Dictionary)

````python
import json
json.load(open('j.json'))
````

### Explanation:

* `json.load()` reads data from a JSON file
* Converts JSON data back into a Python dictionary

### Output:

```python
{'a': 1}
```
