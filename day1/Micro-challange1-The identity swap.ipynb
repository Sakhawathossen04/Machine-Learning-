# Goal : Initialize X = 100 , y = 200 .swap their values so x becomes 200 and y becomes 100.But you do this in exactly one line of code . do not use a temporary variable(like temp = x) 

```python
x = 100
y=200

x,y=y,x

print(f"the current vlaue of x is : {x}")
print(f"the current vlaue of y is : {y}")
```

# What i do in this code?

In c++ or java this is impossible without a temp var. In python ,the syntax x,y=y,x works because:

    1.The right Side y,x creates a temp Tuple object (200,100) in the heap
    2.The left side x,y then reassigning in the index
    3.The temp tuple is then go to grabage
