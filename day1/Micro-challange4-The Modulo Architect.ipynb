# Goal : Input a raw number of seconds .use integer division // and modulo % to calsi exaclty how many hours,minutes,and seconds this represents.

```python
# 1. ইনপুট নেওয়া এবং পূর্ণসংখ্যায় রূপান্তর করা
try:
    raw_sec_str = input("Enter the seconds: ")
    raw_sec = int(raw_sec_str) # **এখানে int() ব্যবহার করা হয়েছে**

    # 2. ঘণ্টা বের করা
    hours = raw_sec // 3600

    sec_after_hours = raw_sec % 3600

    minutes = sec_after_hours // 60

    final_seconds = sec_after_hours % 60
    
    print(f"The time is : {hours} h, {minutes} minutes, {final_seconds} sec")

except ValueError:
    print("Invalid input. Please enter a whole number for seconds.")
```
