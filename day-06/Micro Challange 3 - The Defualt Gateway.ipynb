# Goal : Write a function connect(port = 3306) that prints "Connecting to [port] " . Call it once with no arguments and once with 5432

# Note Default Arguments in Python. These allow you to define a fallback value for a parameter if the caller doesn't provide one.

```python
def connect(port = 3306):
    print(port)

connect()
```

```python
connect(5232)
```

# How it Works
Defining the Default: By writing port=3306 in the function signature, you tell Python: "If the user doesn't give me a value for port, use 3306."

The First Call (connect()): Since no argument is passed, Python "grabs" the stored default from memory and prints: Connecting to 3306.

The Second Call (connect(5432)): The value 5432 overrides the default. Python ignores the 3306 and prints: Connecting to 5432.
