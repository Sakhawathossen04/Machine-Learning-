# Goal : Create a "fake" variable user.age that is calculated from birth year when accessed .
# Deep dive : Use @property.it looks like a variable access.but runs a ethod behind the scenes.This is *Encapsulation*

# Example 1: Thermometer (Celsius → Fahrenheit)

```python
class Thermometer:
    def __init__(self, celsius):
        self.celsius = celsius

    @property
    def fahrenheit(self):
        return (self.celsius * 1.8) + 32
```

```python
t = Thermometer(25)
print(t.fahrenheit)
```

# Example 2: User Age (Birth Year → Age)

```python
from datetime import date

class User:
    def __init__(self,name , birth_year):
        self.name =name
        self.birth_year = birth_year

    @property 
    def age(self):
        current_year = date.today().year
        return current_year - self.birth_year
```

```python
user = User("Sakhawat",2004)
```

```python
user.age
```

# Example 3: Bank Account (Read-only Balance)

```python
class Account:
    def __init__(self):
        self.__balance = 1000

    @property
    def balance(self):
        return self.__balance
```

```python
acc = Account()
print(acc.balance)
```
