# Goal : Find two numbers in a list that add up to Target.
## Deep Dive : Naive : Nested Loops( O(n2)) .Pro : Use a Dictionary .Calculate needed = target - current . If needed is in map ,you win.

## Problem Statement

You are given:

* A list of integers `nums`
* An integer `target`

Your task is to **find two numbers in the list whose sum equals the target**.

---

## Example

```python
nums = [2, 7, 11, 15]
target = 9
```

Here:

```
2 + 7 = 9
```

So the correct pair is `(2, 7)`.


# Example 1: Printing the Numbers

````python
nums = [2,7,11,15]
t = 9
s = {}

for n in nums:
    if t-n in s:
        print(n,t-n)
    s[n] = 1
````

# Example 2: Printing Indices (Interview Favorite)

````python
nums = [3, 2, 4]
t = 6
s = {}

for i, n in enumerate(nums):
    if t - n in s:
        print(s[t - n], i)
    s[n] = i
````
