# Goal : Check if a string is a palindrome (Ignoring space/case ).
## Deep Dive : Clean string --> s==s[::-1] .Sclicing creates a reversed copy in memory .

# Example 1

```python
print('level'=='level'[::-1])
```

# Example 2

```python
print('abc' == 'abc'[::-1])
```

# Example 3

```python
def pal(s):
    return s == s[::-1]

print(pal('madam'))
```
