# Goal : Flatten  [1,[2,[3,4]]] into [1,2,3,4] .
### Deep Dive : Write a recursive function .If item is list -->recurse .Else --> yield.

```python
def flat(l):
 r=[]
 for x in l:
  if isinstance(x,list): r+=flat(x)
  else: r.append(x)
 return r
print(flat([1,[2,[3]]]))
```

```python
print(flat([[1,2],[3,4]]))
```

```python
print(flat([[[1]]]))
```

```python
print(flat([]))
```
