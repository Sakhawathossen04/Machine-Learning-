# Goal : Print 0-100 . Div by 3 "Fizz" , by 5 "Buzz" ,both "FizzBuzz" . 
## Depp Dive : Order matters! Check % 15 (both ) FIRST . If you check 3 first , "FizzBuz" numbers will just print "Fizz."

## Goal

Print numbers from **1 to 100**, but follow these rules:

* If a number is divisible by **3**, print `"Fizz"`
* If a number is divisible by **5**, print `"Buzz"`
* If a number is divisible by **both 3 and 5**, print `"FizzBuzz"`
* Otherwise, print the number itself

---

## Why Is FizzBuzz Important?

FizzBuzz is a classic programming problem that tests:

* Understanding of loops
* Conditional logic
* Correct ordering of conditions

## Example 4: One-Line Solution

```python
for i in range(1, 16):
    print(
        'FizzBuzz' if i % 15 == 0
        else 'Fizz' if i % 3 == 0
        else 'Buzz' if i % 5 == 0
        else i
    )
```
