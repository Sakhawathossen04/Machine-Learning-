# Goal : Create a function power (base ,exp ) .Use functools.partial to create a new function Square(x) that locks exp =2.
# Deep Dive : partials freeeze arguments . this is a useful when you nedd to [ass a function to a callback that expects fewer arguments.

```python
from functools import partial
def power (b,e):
    return b*e

square = partial(power,e=2)
square(5)
```

```python
cube=partial(power,e=3)
print(cube(4))
```
