# Goal : Sort["100px" ,"20x","3px"] numerically (so 3px comes first) .
# Deep dive : Sorted(data,key = lambda x:int(x[:-2]))The key function transforms the item only for comparison purposes,leavning the original data intact

# Example 1 

```python
sorted(["100px","20px","3px"] , key = lambda x : int(x[:-2]))
```

# Example 2

```python
print(sorted(['a','alphabet','cat'],key=len))
```
