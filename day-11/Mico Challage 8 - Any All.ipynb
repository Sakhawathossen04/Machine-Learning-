# Goal : Check if any number in a list in negative. Check if all are positive.
# Deep Dive : These are short circuting operators.any() stops at the first True .all() stops at the first False . This is  O(1) in the best case.

# Example 1

```python
print(any(x<0 for x in [1,2,3,4,5]))
```

# Example 2

```python
print(all(x>0 for x in [1,2,3]))
```
