## What is map()?
map() is a built-in Python function that applies a given function to each item of an iterable (such as a list, tuple, or string).

### Syntax
map(function, iterable)
-function: A transformation function
-iterable: A collection of values

### How map() Works (Conceptually)
map() does three important things:

-Takes one element at a time from the iterable
-Applies the transformation function to it
-Produces a new transformed value

It does NOT modify the original list.

```python
res = map(lambda x : x*x , [1,2,3])
res
```

### map() does not return a list.Instead, it returns a lazy iterator.

```python
#You must convert it to a list:
print(list(res))
```

```python
numbers = [1, 2, 3]
squared = list(map(lambda x: x*x, numbers))

print(numbers)   # Original list
print(squared)   # New transformed list
```
