# Goal : Compare map(lambda...) vs [x ..... for  x ....] .
# Deep Dive : List Comprehensions are generally faster thap map+lambda because they avoid the overhead of calling a python function frame for every single item.

```python
list (map(lambda x : x*x ,range(5)))
```

```python
[x*x for x in range(5)]
```

```python
[x for x in range(10)]
```

```python
[x for x in range(10) if x%2==0]
```
