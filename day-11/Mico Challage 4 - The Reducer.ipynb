# Goal : Calculate the product of list (1*2*3*4) using functools.reduce.
# Deep Dive : Reduce collapses a list into a single value. it tales item 1 and 2 , applies the function , takes the result and item 3 , applies again .. untl one item remains.

# Example 1

```python
from functools import reduce
print(reduce(lambda a,b : a* b ,[1,2,3,4]))
```

# Example 2

```python
from functools import reduce
print(reduce(lambda a,b:a+b,[5,10,15]))
```
