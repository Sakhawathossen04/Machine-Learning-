# Use if else condition in boolean Algebra 

```python
raw_val = input("Enter the score")
score = int(raw_val)
try:
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "A-"
    elif score >= 60:
        grade = "A-"

    else:
        grade = "Fail"


except ValueError:
    print("Invalid")

```
