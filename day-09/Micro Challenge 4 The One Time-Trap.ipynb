## Goal

- Create a generator `g`
- Loop through it **once**
- Then try to loop through the **same generator again**

---

## Deep Dive (The One-Time Trap)

Generators are **exhaustible**.

- When you iterate (loop) through a generator, it produces values **one by one**
- After all values are produced, the generator is **finished**
- The internal “cursor” is now at the end
- You **cannot rewind** a generator   (possible in list)
- To use it again, you must **re-instantiate (create a new generator object)**


```python
def gen():
    yield 1
    yield 2
    yield 3

g = gen()

for x in g:
    print(x)
```

```python
print("Second  Loop:")

for x in g:
    print(x)
```

```python
# Solution this problem we need to re-instantiate

g = gen()

for x in gen():
    print(x)
```
