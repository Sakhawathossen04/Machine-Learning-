## Goal

- Create **two sub-generators** (for example, `gen1` and `gen2`)
- Then create a **main generator** that:
  - Yields all values from `gen1`
  - Then yields all values from `gen2`
- This should be done using **`yield from`**

---

## Deep Dive (Why `yield from` Is Needed)

### What Does `yield from` Mean?

> “Yield all values from another generator or iterable directly to the caller.”

In other words, the main generator **delegates its work** to a sub-generator
instead of running its own loop.

```python
def gen1():
    yield 1
    yield 2

def gen2():
    yield 10
    yield 20

def main():
    yield from gen1()
    yield from gen2()

for v in main():
    print(v)
```
