## Goal

- Create **two generators**:
  - **Generator A**: squares the numbers
  - **Generator B**: keeps only **even numbers** (filters odds)

- Then **chain** them together like this:

```python
filter_even(square(nums))
```

## Deep Dive (Why This Is Important)

This approach creates a **Data Pipeline**:

## source → square → filter_even


In this pipeline:

- Data flows **one item at a time**
- No large intermediate lists are created
- RAM usage remains low
- The process is **memory efficient**

This is a major advantage of generators:  
**no intermediate lists are stored in memory**.
