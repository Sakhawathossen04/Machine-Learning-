## Goal

Manually call `next(gen)` on a generator until it crashes (stops).

---

## Deep Dive: The `next()` Protocol

- A generator produces values **one at a time**
- Each call to `next()` returns the next value
- When the generator has no more values to produce, it raises a  
  **`StopIteration` exception**

This exception means:
> “There are no more items. The generator is finished.”

---

## Example: Manually Calling `next()`


```python
def gen():
    yield 1
    yield 2
    yield 3

g =gen()

print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

# The last next(g) causes a StopIteration error because the generator is exhausted.
