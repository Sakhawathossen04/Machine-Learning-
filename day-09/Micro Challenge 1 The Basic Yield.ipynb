## Goal

Write a function `gen()` that yields **1**, then **2**, then **3** in sequence.  
After that, loop through the generator.

---

## What is `yield`?

- When `return` is used, the function **ends completely** (stops execution).

- When `yield` is used, the function **pauses**, but does not end.  
  It remembers its current position and local variables  
  (this is called the **stack frame** or **state**).

- When you request the next value, the function **continues from where it paused**.


```python
def gen():
    yield 1
    yield 2
    yield 3

for x in gen():
    print(x)

```
