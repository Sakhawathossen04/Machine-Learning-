```python
import numpy as np

# Generate a dummy dataset of 200 centerd samples in 2D
X = np.random.randn(100,2) @ np.array([
    [2,1],
    [1,2]
])

# 1) Compute the 2x2 covariance Matrix of X.

cov_mat = np.cov(X,rowvar = False)

# 2) Find the eigenvalues and eigenvectors of the covariance matrix.

evals , evec = np.linalg.eig(cov_mat)

# 3) Identift the principle eigenvector(the one associated with the largest eigenvalue)

idx_max = np.argmax(cov_mat)
principal_eigenvector = evec[:, idx_max] 

# 4)Project X onto the principle eigenvector using the dot product

X_projected = (X @ principal_eigenvector).reshape(-1, 1)  # shape (100,1)
```

```python
print("Covariance matrix:\n", cov_mat, "\n")
print("Eigenvalues:\n", evals, "\n")
print("Principal eigenvector:\n", principal_eigenvector, "\n")
print("Projected X shape:", X_projected.shape)
print("First 5 projected values:\n", X_projected[:5])
```
