```python
import numpy as np

S =np.array([
    [3,1],
    [1,3,]
])

evals , evec = np.linalg.eig(S)

v1 = evec[:,0]
v2 =evec[:,1]

Check = v1 @ v2


print("Problem 3 Results:")
print("Eigenvalues:", evals)
print("Eigenvectors (columns):\n", evec)
print(f"Dot product of eigenvectors: {Check}")
print("If the result is ~0, they are perfectly orthogonal.")
```
