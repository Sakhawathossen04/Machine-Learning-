# Method:1

```python
import numpy as np

A = np.array([
    [0.8 , 0.3],
    [0.2,0.7],
]
)
K=10
A_k_derect_power = np.linalg.matrix_power(A,K)
```

# Method:2  Eigendecomposition (Q * Lambda ^k * Q^ -1)

Find eigenvalues and eigenvectors of A

```python
eigenvalues , Q = np.linalg.eig(A)

lambda_k = np.diag(eigenvalues ** K )

Q_inv = np.linalg.inv(Q)

q_lambda_k_q_1 = Q @ lambda_k @ Q_inv

```

```python
print("Problem 1 Results:")
print(f"Direct Power:\n{A_k_derect_power}\n")
print(f"Eigendecomposition Power:\n{A_k_derect_power}\n")

# Compare (prove identical up to numerical precision)
diff = A_k_derect_power - q_lambda_k_q_1
print("Diff matrix:\n", diff)
```
