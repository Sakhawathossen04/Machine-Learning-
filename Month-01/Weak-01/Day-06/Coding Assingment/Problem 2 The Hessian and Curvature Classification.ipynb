```python
import numpy as np

# The hessian and curvature classification

def classify_critical_point(hessian_matrix):

    # Compute the eigenvalues of the hessian_matrix

    eigenvalues = np.linalg.eigvals(hessian_matrix)


    # Implement the classification logic
    if np.all(eigenvalues > 0):
        classification = "Local Minimum"

    elif np.all(eigenvalues < 0) :
        classification = "Local Maximum"

    else:
        classification = "saddle point"

    return eigenvalues ,classification

# Define the hessian matrix

H_A = np.array([
    [4,1],
    [1,3],
])

H_B = np.array([
    [-2,0],
    [0,-5]
])

H_C =np.array([
    [2,4],
    [4,-1],
])



print("Problem 2 results:")

evals_A , result_A = classify_critical_point(H_A)
print(f"point A : EigenValues {evals_A.round(2)} -- > {result_A}")

evals_B, result_B = classify_critical_point(H_B)
print(f"Point B: Eigenvalues {evals_B.round(2)} -> {result_B}")

evals_C, result_C = classify_critical_point(H_C)
print(f"Point C: Eigenvalues {evals_C.round(2)} -> {result_C}")
```
