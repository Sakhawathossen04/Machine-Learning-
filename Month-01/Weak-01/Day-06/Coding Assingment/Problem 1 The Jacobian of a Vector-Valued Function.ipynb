```python
import sympy as sp
import numpy as np

#problem: The jacobian of a vector-valued function 

x1,x2 = sp.symbols('x1 x2')

# Define the output vector functions
y1 = x1**2 + sp.sin(x2)
y2 = x1* x2
y3 = sp.exp(x1)+2*x2


# Create sympy matrices
Y = sp.Matrix([y1,y2,y3])
X = sp.Matrix([x1,x2])

#Compute the symbolic jacobian matric of Y with repect to X

jacobian_sym = Y.jacobian(X)


# Evaluate the jacobian at the point(x1=1 , x2 =0)
jacobian_num = jacobian_sym.subs({x1:1,x2:0})

print("Problem 1 Results:")
print(f"Symbolic Jacobian:]\n{jacobian_sym}\n")
```
