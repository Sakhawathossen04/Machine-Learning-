```python
import numpy as np

# We have 2 equation:
# 3x1 + 2y1 = 28
# 1x2 + 4y2 = 16

# Now we wanna represent coeffsX = b

coeffs = np.array([
    [3, 2],
    [1, 4]
], dtype=float)

b = np.array([28, 16], dtype=float)

prices = np.linalg.solve(coeffs, b)

print("--- Problem 3 Results ---")
print(f"GPU Price per hour : ${prices[0]:.2f}")
print(f"CPU Price per hour : ${prices[1]:.2f}")
 
```
