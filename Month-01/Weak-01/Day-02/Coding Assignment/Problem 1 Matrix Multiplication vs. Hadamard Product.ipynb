```python
import numpy as np

A = np.array([
    [1,2],
    [3,4],
])

B = np.array([
    [0,1],
    [-1,0],
])

AB = A @ B 
BA = B @ A

# Handamard product(A_B) A ,B 
A_B = A*B

print (" --- Problem 1 Results ---")
print (f"AB :\n{AB }\n")
print (f"BA :\n{BA }\n")
print (f" Element - wise ( Hadamard ):\n{ A_B }\n")
```
