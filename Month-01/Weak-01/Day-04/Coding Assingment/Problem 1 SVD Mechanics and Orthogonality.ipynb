```python
import numpy as np

# Problem 1: SVD Mechanics and Orthogonality
np.random.seed(42)
X_prob1 = np.random.rand(5, 3)

# Compute full SVD of X_prob1
U, Sigma_vals, Vt = np.linalg.svd(X_prob1, full_matrices=True)

# Reconstruct X_prob1 from U, Sigma_vals, and Vt
Sigma = np.zeros((5, 3))
np.fill_diagonal(Sigma, Sigma_vals)
X_reconstructed = U @ Sigma @ Vt

# Compute U^T @ U to verify orthogonality
U_ortho_check = U.T @ U

print("Problem 1 Results:")
print("\nOriginal X:")
print(X_prob1)

print("\nU shape:", U.shape)
print("Sigma_vals:", Sigma_vals)
print("Vt shape:", Vt.shape)

print("\nReconstructed X:")
print(X_reconstructed)

print("\nReconstruction error:")
print(np.linalg.norm(X_prob1 - X_reconstructed))

print("\nU^T U:")
print(U_ortho_check)
```
