```python
import numpy as np
# Problem 2: Low-Rank Approximation
A = np.random.rand(6, 4)
k = 2

# Compute SVD of A
U_A, Sigma_A, Vt_A = np.linalg.svd(A, full_matrices=False)

# Truncate matrices to keep only top k singular values/vectors
U_k = U_A[:, :k]
Sigma_k_matrix = np.diag(Sigma_A[:k])
Vt_k = Vt_A[:k, :]

# Compute the rank-k approximation (A_k)
A_k = U_k @ Sigma_k_matrix @ Vt_k

# Compute the Frobenius norm of the difference (Information Loss)
info_loss = np.linalg.norm(A - A_k, ord='fro')

print("Problem 2 Results:")
print("\nOriginal A:")
print(A)

print("\nSingular values:")
print(Sigma_A)

print("\nRank-2 approximation A_k:")
print(A_k)

print("\nInformation loss ||A - A_k||_F:")
print(info_loss)
```
