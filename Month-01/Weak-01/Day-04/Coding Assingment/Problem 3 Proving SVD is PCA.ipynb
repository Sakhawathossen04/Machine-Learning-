```python
import numpy as np

# Problem 3: Proving SVD is PCA
np.random.seed(42)
X_prob3 = np.random.rand(100, 3)

# Step 1: Mean-center the data (subtract column means)
X_centered = X_prob3 - np.mean(X_prob3, axis=0)

# Step 2: Compute SVD of centered data and extract Vt
U_pca, Sigma_pca, Vt_pca = np.linalg.svd(X_centered, full_matrices=False)

# Step 3: Compute covariance matrix of centered data
cov_matrix = np.cov(X_centered, rowvar=False)

# Step 4: Compute eigenvalues and eigenvectors of covariance matrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Sort eigenvalues/eigenvectors in descending order
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Compare rows of Vt with transpose of eigenvectors
comparison = np.allclose(np.abs(Vt_pca), np.abs(eigenvectors.T))

print("Problem 3 Results:")

print("\nOriginal Data Shape:")
print(X_prob3.shape)

print("\nColumn means before centering:")
print(np.mean(X_prob3, axis=0))

print("\nColumn means after centering (should be close to 0):")
print(np.mean(X_centered, axis=0))

print("\nSingular values from SVD:")
print(Sigma_pca)

print("\nVt from SVD:")
print(Vt_pca)

print("\nCovariance matrix:")
print(cov_matrix)

print("\nEigenvalues of covariance matrix (sorted descending):")
print(eigenvalues)

print("\nEigenvectors of covariance matrix (sorted columns):")
print(eigenvectors)

print("\nTranspose of eigenvectors:")
print(eigenvectors.T)

print("\nAre rows of Vt equal to covariance eigenvectors (up to sign)?")
print(comparison)
```
