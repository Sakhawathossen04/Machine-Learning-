In machine learning, if one feature can be perfectly predicted by a linear combination of other
features, your matrix loses full rank (it becomes singular). This breaks algorithms that rely on
matrix inversion, like Ordinary Least Squares (OLS) regression.

You are given a dataset matrix X where rows are samples and columns are features: x1 (tem-
perature in Celsius), x2 (humidity), and x3 (temperature in Fahrenheit). Prove computationally

that this feature set is linearly dependent by finding the rank of the matrix.

```python
import numpy as np

dataset =np.array([
    [1,20,45,68],
    [1,25,50,77],
    [1,30,60,86],
    [1,15,30,59]
])

matrix_rank = np.linalg.matrix_rank(dataset)
# features are independent iff rank == number of columns

is_independent = (matrix_rank == dataset.shape[1])

print("Problem 2 Resilt:")
print(f"Matrix Rank: {matrix_rank}")
print(f"Are Features Linearly Independent? --> {is_independent}")
```
