Write a generic function that takes an integer n and generates the standard basis B for Rn as
a list of NumPy arrays.

```python
import numpy as np

def generate_standard_basis(n):
    basis_vector = []
    for i in range(n):
        v = np.zeros(n)
        v[i]=1
        basis_vector.append(v)

    return basis_vector
```

```python
n = 4
basis = generate_standard_basis(n)
```
