```python
import numpy as np

# Point (x,y)
x_val = 2.0
y_val = 1.0

h = 1e-5

# Multivariable function
def f(x, y):
    return 3*(x**2)*y + np.sin(x)*np.exp(y)

# Partial derivative w.r.t x
def approx_df_dx(func, x, y, h=1e-5):
    return (func(x+h, y) - func(x, y)) / h

# Partial derivative w.r.t y
def approx_df_dy(func, x, y, h=1e-5):
    return (func(x, y+h) - func(x, y)) / h


df_dx = approx_df_dx(f, x_val, y_val)
df_dy = approx_df_dy(f, x_val, y_val)

print (" Problem 1 Results :")
print (f" Numerical df/dx at (2.0 , 1.0) : { df_dx }")
print (f" Numerical df/dy at (2.0 , 1.0) : { df_dy }\n")
```
