# Goal : Create @admin_required . if global USER != 'admin ' ,raise permissionError .Deep Dive : The decorator acts as a gatekeeper . it executes before the sensitive function is entered . if the check fails, the stack frame for the target function is never even created.

# Example 1: Admin check

```python
USER = "admin"

def admin_required(func):
    def wrap():
        if USER != "admin":
            raise PermissionError
        return func()
    return wrap

@admin_required
def dashboard():
    print("Admin Panel")

dashboard()
```

# Example 2: Non-admin

```python
USER = "guest"

def admin_req(func):
    def wrap():
        if USER !="admin":
            print("Acces Denied")
            return
        func()
    return wrap

@admin_req
def dashboard():
    print("Admin")

dashboard()
```

# Example 3: With args

```python
USER = "admin"

def guard(func):
    def wrap(*args):
        if USER != "admin":
            return "Denied"
        return func(*args)
    return wrap

@guard
def delete(item):
    return f"{item} deleted"

print(delete("file"))
```
