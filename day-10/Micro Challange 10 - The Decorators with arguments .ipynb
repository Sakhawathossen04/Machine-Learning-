# Example 1: Repeat

```python
def repeat(times):
    def deco(func):
        def wrap():
            for _ in range(times):
                func()
        return wrap
    return deco

@repeat(3)
def hello():
    print("Hi")

hello()
```

```python
def multiply(k):
    def deco(func):
        def wrap():
            return func() * k
        return wrap
    return deco

@multiply(5)
def value():
    return 2

print(value())
```

# Example 3: Parameterized log

```python
def log(msg):
    def deco(func):
        def wrap():
            print(msg)
            return func()
        return wrap
    return deco

@log("Running")
def task():
    print("Task")

task()
```
