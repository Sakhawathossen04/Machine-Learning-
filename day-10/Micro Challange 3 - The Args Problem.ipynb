# Goal: Try to decorate afunction that takes arguments add(a,b) with wrapper that tales none,Deep Drive : It crashes . The inner wrapper function MUST accept *args and  **kwargs to be compatible wth any target function signature

# Example 1: Positional args

```python
def deco(func):
    def wrap(*args):
        print("args:" ,args)
        return func(*args)
    return wrap

@deco
def add(a,b):
    return a+b

print(add(2,3))
```

# Example 2: Keyword args

```python
def deco(func):
    def wrap(**kwargs):
        print(kwargs)
        return func(**kwargs)
    return wrap
@deco
def show(name="X"):
    return name

print(show(name="Sakhawat"))
```

# Example 3: Both args

```python
def deco(func):
    def wrap(*args,**kwargs):
        return func(*args,**kwargs)
    return wrap

@deco
def mul(a,b):
    return a+b
@deco
def show(name = "X"):
    return name 

print(mul(4,5))
print(show(name="Sakhawat"))
```
