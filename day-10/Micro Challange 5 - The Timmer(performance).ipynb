# Goal : Create a @ timmer decorator that prints execution time using time.time() .Deep Dive:This is AOP .We inject timing logic into the function without modifiying the functions actual code.

```python
import time
def timer(func):
    def wrapper():
        start_time = time.time()

        func()
        
        end_time = time.time()
        print(f"Execution time:: {end_time - start_time} seconds")
 
    return wrapper

@timer
def say_hi():
    print("say_hi")

say_hi()
```

# Example 1: Simple timer

```python
import time

def timer(func):
    def wrap():
        start = time.time()
        func()
        print(time.time() - start)
    return wrap

@timer
def work():
    time.sleep(1)

work()
```

# Example 2: Timer with args

```python
import time

def timer(func):
    def wrap(*args):
        start = time.time()
        result = func(*args)
        end = time.time()
        print(f"Execution time:: {end - start} seconds")
        return result
    return wrap

@timer
def slow_add(a,b):
    time.sleep(3)
    return a+b

print(slow_add(3,4))
    
```

# Example 3: Loop timing

```python
import time

def timer(func):
    def wrap():
        s = time.time()
        func()
        print("Elapsed:", time.time()-s)
    return wrap

@timer
def loop():
    for _ in range(1000000):
        pass

loop()
```
