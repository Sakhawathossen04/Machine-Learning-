# Goal : Write a function wrapper(func) that runs code before/after func.apply it manually : new_func = wrapper(old_func) .Deep Drive : This demystifies the @ syntax.A decorator is simply a function that takes a function and returns a function.

```python
# Example 1 : Basic Wrapper 
```

```python
def wrapper(func):
    def inner():
        print("Beffor")
        func()
        print("After")

    return inner

def hello():
    print("Hi there")

new_func = wrapper(hello)
new_func()
```

```python
# Example 2 : Logging wrapper
```

```python
def logger(func):
    def inner():
        print("Calling function")
        func()
    return inner

def work():
    print("Workining")

logger(work)()
```
