# Hash Lookup (O(1))  Easy Explanation

This note explains **Hash Lookup (O(1))**.  
It shows how we can quickly check whether a value exists in a **Set or Dictionary**.

---

## What is the problem?

### Task
Check whether a value (called **target**) exists inside a **Set or Dictionary**.

---

## Why do we use Hash?

- Searching in a **List / Array** takes **O(n)** time  
  (we check items one by one)

- Searching in a **Set / Dictionary** usually takes **O(1)** time  
  (very fast)

**Reason:** Hash Function

---

## Pseudo Code Explanation (Step by Step)

### Step 1: Calculate hash value

hash_value = calculate_hash(target)



- Every value is converted into a number  
- This number is called a **hash**
- Hash is created by a **hash function**
- Same input usually gives the same hash

**Example:**  
`"apple"`  hash = `94573` (example only)

---

## Step 2: Find Bucket Index
```text
bucket_index = hash_value % set.total_buckets

Memory is divided into small parts called buckets

Modulo (%) decides which bucket will store the value

Example:

-total_buckets = 10
-hash_value = 94573
-bucket_index = 94573 % 10 = 3

So, the value goes to bucket 3.

## Step 3: Direct Memory Jump
```text
memory_slot = set.buckets[bucket_index]

-Program jumps directly to bucket 3
-No need to check other buckets
-This direct access is the reason for O(1) time complexity

## Step 4: Check Target in the Bucket

IF memory_slot contains target:
    RETURN True
ELSE:
    RETURN False

Only the selected bucket is checked

Found  True

Not found  False

## Full Process in One Line

Hash the value  find the bucket  direct jump  compare

## What is Collision?

Sometimes two different values go to the same bucket
This situation is called a collision

In that case:
Bucket stores values in a small list or structure
With a good hash function, collisions are rare
Average time complexity still remains O(1)


## Python Example

s = {10, 20, 30}

print(20 in s)  # True
print(25 in s)  # False


## Quick Memory Trick

List  Search = O(n)

Set / Dictionary  Search = O(1)

````python
def set_contains(my_set, target):
    # Step 1: Compute mathematical signature (hash)
    hash_value = hash(target)

    # Step 2: Calculate memory address (bucket index)
    total_buckets = len(my_set)
    bucket_index = hash_value % total_buckets

    # Step 3: Direct jump to bucket
    buckets = set.buckets[bucket_index]

    # (Assume elements are already placed into buckets)
    for value in my_set:
        index = hash(value) % total_buckets
        buckets[index].append(value)

    memory_slot = buckets[bucket_index]

    # Step 4: Check target
    if target in memory_slot:
        return True
    else:
        return False
````

````python
s = {10, 20, 30}

print(set_contains(s, 20)) 
print(set_contains(s, 25)) 
````
