# Error Handling (Exceptions)

When an error occurs . it bubbles up the call stack . if nnothing catches it,the program craches .

Defensive Programming : We use try/except block to catch these bubbles. This is required for Data Science pipelines where one bad row of data shoildn't stop a 10 h tranning process. 

```python
while True:
    try:
        val = int(input("Enter the value:"))
        print(100/val)
        break
    except ValueError:
        print("The text is not allowed")
    
    except ZeroDivisionError:
        print ("Cannot dividde by zero!!")
    
    
```
