# Goal :  Write a try/except block that divides 2 numbers . add a finnaly block that print "Execution Complete" regardless of whether the division succeeded or failed.

```python
try:
    x = int(input("Enter the value of X: "))
    y = int(input("Enter the value of Y: "))
    print(f"The value of division: {x / y}")

except ZeroDivisionError:
    print("Division by zero is not allowed (infinity).")

except ValueError:
    print("Invalid input! Please enter integers only.")

finally:
    print("The division attempt has finished (success or error).")
```
